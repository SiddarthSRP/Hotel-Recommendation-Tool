```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
import re
url = 'https://www.booking.com/searchresults.html?ss=Chennai%2C+India&ssne=Ooty&ssne_untouched=Ooty&efdco=1&label=gen173nr-1FCAQoggJCDWNpdHlfLTIxMTQ4ODhIM1gEaGyIAQGYATG4ARfIAQzYAQHoAQH4AQOIAgGoAgO4At--3K8GwAIB0gIkMGRhNWE1ZmItY2I2My00Y2MyLWI1YWItMjQ2ZGVhMzFhM2Rm2AIF4AIB&aid=304142&lang=en-us&sb=1&src_elem=sb&src=searchresults&dest_type=city&group_adults=1&no_rooms=1&group_children=0'
headers = {
    'User-Agent': 'Mozilla/5.0 (X11; CrOS x86_64 8172.45.0) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/51.0.2704.64 Safari/537.36',
    'Accept-Language': 'en-US, en;q=0.5'
}
response = requests.get(url, headers=headers)

# Check if the request was successful
if response.status_code == 200:
    # Parse the HTML content of the page
    soup = BeautifulSoup(response.text, 'html.parser')

    # Find all the hotel elements in the HTML document
    hotels = soup.findAll('div', {'data-testid': 'property-card'})

    hotels_data = []

    # Loop over the hotel elements and extract the desired data
    for hotel in hotels:
        # Extract the hotel name
        name_element = hotel.find('div', {'data-testid': 'title'})
        name = name_element.text.strip() if name_element else None

        # Extract the hotel location
        location_element = hotel.find('span', {'data-testid': 'address'})
        location = location_element.text.strip() if location_element else None

        # Extract the hotel rating
        rating_element = hotel.find('div', {'class': 'a3b8729ab1 d86cee9b25'})
        rating_text = rating_element.text.strip() if rating_element else None

        # Extract only the numerical part from the rating text using regular expressions
        if rating_text:
            rating_match = re.search(r'\d+\.\d+', rating_text)
            rating = rating_match.group() if rating_match else None
        else:
            rating = None

        # Extract the hotel URL
        link_element = hotel.find('a', class_='a78ca197d0')
        if link_element:
            hotel_url = link_element['href']
            
            response_hotel = requests.get(hotel_url)
            if response_hotel.status_code == 200:
                soup_hotel = BeautifulSoup(response_hotel.content, 'html.parser')
                review_elements = soup_hotel.findAll('div', {'data-testid': 'featuredreview-text'})

                reviews = [review_element.text.strip() for review_element in review_elements]

                hotels_data.append({
                    'name': name,
                    'location': location,
                    'rating': rating,
                    'reviews': reviews
                })

    df = pd.DataFrame(hotels_data)
    print(df)
    # Save the DataFrame to a CSV file
    df.to_csv('hotels_data.csv', index=False)

else:
    print("Failed to retrieve data from Booking.com")
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from nltk.sentiment.vader import SentimentIntensityAnalyzer

sid = SentimentIntensityAnalyzer()

# Read the CSV file
df = pd.read_csv('hotels_data.csv')

df['sentiment_scores'] = df['reviews'].apply(lambda x: sid.polarity_scores(x))

df['compound_score'] = df['sentiment_scores'].apply(lambda x: x['compound'])

df['sentiment'] = df['compound_score'].apply(lambda x: 'positive' if x >= 0.06 else ('negative' if x <= 0 else 'neutral'))

ranked_hotels = df.sort_values(by=['compound_score'], ascending=False)

print(ranked_hotels[['name', 'compound_score', 'sentiment']])

plt.figure(figsize=(12, 6))

plt.subplot(1, 2, 1)
plt.barh(ranked_hotels['name'], ranked_hotels['compound_score'], color='red')
plt.xlabel('Compound Sentiment Score')
plt.ylabel('Hotel Name')
plt.title('Compound Sentiment Score of Hotels')
plt.gca().invert_yaxis()
plt.grid(axis='x', linestyle='--', alpha=0.9)

plt.tick_params(axis='both', which='major', labelsize=10)

plt.subplot(1, 2, 2)
sentiment_counts = df['sentiment'].value_counts()
plt.pie(sentiment_counts, labels=sentiment_counts.index, autopct='%1.1f%%', colors=['lightgreen', 'lightcoral', 'lightskyblue'])
plt.title('Sentiment Distribution')

plt.tight_layout()
plt.savefig('sentiment_analysis.jpg')
plt.show()
```
